>

Um `programa Python` que utilize `expressões regulares` pode ajudar a identificar vários **tipos de erros** comuns em **livros**, especialmente erros de formatação e problemas relacionados à consistência do texto. Aqui estão alguns exemplos de erros comuns que podem ser identificados usando expressões regulares:

* Erros de gramática e ortografia: erros de digitação, concordância verbal e nominal, uso incorreto de pontuação e outros erros gramaticais.

* Problemas de formatação: você pode usar expressões regulares para encontrar erros de formatação, como espaços em excesso, tabulações inadequadas ou alinhamentos inconsistentes.

* Abreviações e acrônimos: você pode usar expressões regulares para encontrar abreviações ou acrônimos que não foram definidos ou explicados anteriormente no texto.

* Citações e referências: expressões regulares podem ser úteis para localizar citações ou referências que precisam de formatação especial.

* OUTROS TIPOS DE ERROS: não considerem apenas os tipos de erros citados acima.


**IMPORTANTE:** Lembre-se de que expressões regulares podem ser poderosas, mas também complexas. Dependendo da complexidade dos erros que você deseja identificar, pode ser necessário ajustar as expressões regulares de acordo com as características específicas do seu texto. Além disso, é importante ter em mente que as expressões regulares podem não ser a melhor ferramenta para todos os tipos de erros em livros, especialmente problemas mais contextuais ou semânticos, que podem exigir abordagens de PLN mais avançadas.

**Código**

```
import requests
from bs4 import BeautifulSoup
import re
import enchant

# Função para obter o conteúdo HTML da URL
def baixar_html(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.content
    else:
        return None

# Função para verificar erros de gramática e ortografia
def checar_erros_gramaticas_ortograficos(text):
    # Palavras que são nomes de autores, acrônimos ou abreviações técnicas
    palavras_reservadas = [
        # "dfrac", "et", "al", "endofóricas", "Faraco", "Roncarati", "PLN", "Milidiú", "Vanin", "Bick", "Ferradeira"
    ]

    # Dicionário da língua portuguesa
    d = enchant.Dict("pt_BR")
    # Dicionário da língua inglesa
    d_ingles = enchant.Dict("en_US")

    erros = []

    # Exemplo de verificação de erros de concordância verbal:
    erros_concordancia_verbal = re.findall(r'\b(\w+) (é|são) (\w+) (o|a|os|as) \w+\b', text)
    for palavra in erros_concordancia_verbal:
        erros.append(f"Erro de concordância verbal: {palavra}")

    erros_concordancia_nominal = re.findall(r'\b(\w+) (?:é|são) (?:\w+ (?:os|as)? )?\w+ (?:os|as) \w+\b', text)
    for palavra in erros_concordancia_nominal:
        erros.append(f"Erro de concordância nominal: {palavra}")

    erros_acentuacao = re.findall(r'\b\w*[aeiou](?:[áàâãéèêíóôõú])\w*\b', text)
    for palavra in erros_acentuacao:
        # Verifica no dicionário se realmente é um erro de acentuação
        if not d.check(palavra) and palavra not in palavras_reservadas:
          erros.append(f"Erro de acentuação: {palavra}")

    # Exemplo de uso incorreto de pronomes (uso incorreto de "mim" e "eu"):
    erros_pronome = re.findall(r'\b(eu|mim) (e|é) (ele|ela)\b', text)
    for palavra in erros_pronome:
        erros.append(f"Erro pronomial: {palavra}")

    # Exemplo de verificação de uso incorreto de apóstrofos:
    erros_apostrofe = re.findall(r"\b'\w+", text)
    for palavra in erros_apostrofe:
        erros.append(f"Apóstrofe mal utilizado: {palavra}")

    # Exemplo de verificação de repetições de letras:
    erros_letras_repetidas = re.findall(r'\b(\w)\1{2,}\b', text)
    for palavra in erros_letras_repetidas:
        erros.append(f"Letra repetida: {palavra}")

    # Use expressões regulares para encontrar palavras
    palavras = re.findall(r'\b\w+\b', text)

    for palavra in palavras:
        # Verifica se a palavra não está no português, não é uma referência em ingles e não é uma variável que contém "_"
        if not d.check(palavra) and not d_ingles.check(palavra) and not re.findall(r"_", palavra) and palavra not in palavras_reservadas:
              erros.append(f"Palavra não encontrada no dicionário: {palavra}")

    if len(erros) > 0:
      print("Erros de gramática e ortografia:")
      for error in erros:
          print(error)

    return len(erros)

# Função para encontrar erros de formatação
def checar_espacos_excesso(text):
    # Exemplo: Encontrar espaços em excesso
    espacos_excesso = re.findall(r'\s{2,}', text)

    return len(espacos_excesso)

# Função para detectar citações e referências
def checar_citacoes(text):
    citacoes = []
    # Exemplo de detecção de citações entre aspas
    citacao = re.findall(r'"(.*?)"', text)
    for match in citacao:
        citacoes.append(f"Citação incompleta: {match}")
        print(f"Citação incompleta: {match}")
    return len(citacoes)

# Função principal
def main():
    urls = [
      "https://brasileiraspln.com/livro-pln/1a-edicao/parte6/cap12/cap12.html",
      "https://brasileiraspln.com/livro-pln/1a-edicao/parte6/cap13/cap13.html"
    ]

    total_erros = 0

    for url in urls:
      html = baixar_html(url)

      if html is not None:
          capitulo = "12" if url == "https://brasileiraspln.com/livro-pln/1a-edicao/parte6/cap12/cap12.html" else "13"

          print(f"============== CAPÍTULO {capitulo} ==============")

          # Analisar o conteúdo HTML
          soup = BeautifulSoup(html, 'html.parser')

          tags = soup.find_all(['section', 'p']) if capitulo == "12" else soup.find_all(['main', 'p'], attrs={'class': ['content']})

          secao_do_texto = 0

          total_erros_gramaticais_ortograficos = 0
          total_citacoes_erradas = 0
          total_espacos_excesso = 0

          for tag in tags:
              secao_do_texto += 1
              text_content = tag.text.strip()

              print(f"(Parágrafo {secao_do_texto})")
              erros_gramaticais_ortograficos = checar_erros_gramaticas_ortograficos(text_content)
              citacoes_erradas = checar_citacoes(text_content)
              espacos_excesso = checar_espacos_excesso(text_content)

              total_erros_gramaticais_ortograficos += erros_gramaticais_ortograficos
              total_citacoes_erradas += citacoes_erradas
              total_espacos_excesso += espacos_excesso

          total_erros_capitulo = total_erros_gramaticais_ortograficos + total_citacoes_erradas + total_espacos_excesso
          total_erros += total_erros_capitulo

          print(f"Erros gramaticas e ortográficos: {total_erros_gramaticais_ortograficos}")
          print(f"Citações erradas: {total_citacoes_erradas}")
          print(f"Espaços em excesso(+2): {total_espacos_excesso}")
          print(f"Erros totais no capítulo {capitulo}: {total_erros_capitulo}")

      else:
          print(f"Não foi possível obter o conteúdo HTML de {url}")

    print("==================================================================")
    print(f"Total de erros {total_erros}")

if __name__ == "__main__":
    main()
```
